#Mapa de estaciones de Mibici

```python
import plotly.graph_objects as go
```

```python
nomen_activas = nomen[nomen.status == 'IN_SERVICE']
# Crear el gráfico de puntos en el mapa
fig = go.Figure()

# Agregar los puntos al mapa
fig.add_trace(go.Scattermapbox(
    lat=nomen_activas['latitude'],
    lon=nomen_activas['longitude'],
    mode='markers',
    marker=go.scattermapbox.Marker(
        size=10,
        color='green',  # Color de los puntos
        opacity=0.8
    ),
    text=nomen_activas['id'],  # Etiquetas de los puntos
    hoverinfo='text'
))

# Configurar diseño del mapa
fig.update_layout(
    title='Estaciones mi Bici en Jalisco',
    autosize=True,
    hovermode='closest',
    mapbox=dict(
        style='open-street-map',  # Estilo del mapa
        zoom=11,  # Nivel de zoom inicial
        center=dict(
            lat=20.6767,  # Latitud centrada en Jalisco
            lon=-103.3475  # Longitud centrada en Jalisco
        )
    )
)

# Mostrar el mapa
fig.show()
```

# ¿Cuantos viajes se hacen?

```python
year = lambda y : y[0:4]
df['Year'] = df.Trip_start.apply(year)
```

```python
Conteo = df.groupby(['Year']).agg({'Year': 'count'}).rename(columns={'Year':'Count'}).reset_index()
```

```python
fig = px.bar(Conteo, x = 'Year', y = 'Count')
fig.update_traces(marker_color='green')
fig.update_layout(
    yaxis_title = 'Número de viajes',
    xaxis_title = 'Años',
    title_text = 'Viajes en Mibici',
    template = 'presentation'
)
fig
```

# Grafica de Pastel, según el sexo

```python
Genero = df.groupby(['Sex']).User_Id.nunique().reset_index()
```

```python
Genero
```

```python
fig = px.pie(Genero, values='User_Id', names='Sex', color_discrete_sequence=px.colors.sequential.Plasma)
fig.update_layout(
    title_text = 'Distribución de Genero',
    template = 'presentation'
)
fig.show()
```

# Vista por edades

```python
fig = px.pie(Genero, values='Age', names='Sex', color_discrete_sequence=px.colors.sequential.Plasma)
fig.update_layout(
    title_text = 'Distribución de Genero',
    template = 'presentation'
)
fig.show()
```
